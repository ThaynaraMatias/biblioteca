```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
def estado_info(estado: str) -> dict:
  print(f'...Pegando as informações {estado}')
  estado_url = f'https://www.ibge.gov.br/cidades-e-estados/{estado}.html'
  pag = requests.get(estado_url)

  soup = BeautifulSoup(pag.content, 'html.parser')
  indicadores = soup.select('.indicador')

  estado_dict = {
        ind.select('.ind-label')[0].text: ind.select('.ind-value')[0].text
        for ind in indicadores
    }

  estado_dict['Estado'] = estado

  return estado_dict

estado_info('ce')
```

```python
estados = ['AC', 'AL', 'AP', 'AM', 'BA', 'CE', 'DF', 'ES', 'GO', 'MA', 'MT', 'MS', 'MG', 'PA', 'PB', 'PR', 'PE', 'PI', 'RJ', 'RN', 'RS', 'RO', 'RR', 'SC', 'SP', 'SE', 'TO']
estados_data = [estado_info(estado) for estado in estados]
```

```python
df = pd.DataFrame(estados_data)
df.head()
#df.info()
```

```python
df.columns
```

```python
df = df.replace({
    '\.': '',
    '\,': '.',
    'matrícula':'',
    '\[\d+\]': '',
    ' hab/km²': '',
    ' km²': '',
    ' pessoas': '',
    'R\$.*': '',
    ' veículos': ''
}, regex=True)

num_cols = ['População residente', 'Área Territorial', 'Matrículas no ensino fundamental', 'IDH Índice de desenvolvimento humano', 'Rendimento mensal domiciliar per capita', 'Total de veículos', 'Despesas empenhadas', 'Receitas realizadas']

df[num_cols] = df[num_cols].apply(lambda x: x.str.strip())
df[num_cols] = df[num_cols].apply(pd.to_numeric)

df.head()
```
